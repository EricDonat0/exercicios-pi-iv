Faça este cálculo usando Python. Quantas interações são necessárias para que tenhamos precisão de 6 casas decimais?

$$
e = \sum^{\infty}_{n=0} = \frac{1}{n!} = 1 + \frac{1}{1}+ \frac{1}{1 \cdot 2} + \frac{1}{1\cdot2\cdot3} + \dots
$$

```python
import math

def calcular_soma(n_max):
  euler = 0

  for n in range(0, n_max+1):
    euler += 1/math.factorial(n)

  return euler
```

# Precisão

```python
erro_min = 0.000001
n_teto = 50
n_inicial = 0
passo = 1

euler_antigo = 0
euler = 0
erro = 0
passos = 0
cont = 0

n_s = []
euler_s = []
erros = []

for n in range(n_inicial, n_teto, passo):
  euler_atual = calcular_soma(n)
  erro_parcial =  euler_atual - euler_antigo
  cont += 1

  n_s.append(n)
  erros.append(erro_parcial)
  euler_s.append(euler_atual)

  print(f"Iteração {cont}, Euler: {euler_atual}, Euler Antigo: {euler_antigo}, Erro: {erro_parcial}")

  if erro_parcial <= erro_min:
    erro = erro_parcial
    euler = euler_atual
    passos = n
    break
  else:
    euler_antigo = euler_atual

print(f"\nResultado final")
print(f"Passo {passos}, Euler: {euler}, Erro: {erro}")
```
