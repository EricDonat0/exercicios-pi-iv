# Secante


$$
f(x) = \sec(x) = \frac{1}{\cos(x)}
$$



### Derivada

$$
f'(x) = \frac{d}{dx} \left( \frac{1}{\cos x} \right)
       = \frac{0 \cdot \cos x - 1 \cdot (-\sin x)}{\cos^2 x}
       = \frac{\sin x}{\cos^2 x}
       = \sec(x) \tan(x)
$$


### Derivadas de sec(x) avaliadas em (x=0)

$$
f(x) = \sec(x) = \frac{1}{\cos x}
$$

$$
f'(x) = \sec(x) \tan(x), \quad f'(0) = 0
$$

$$
f''(x) = \sec(x) (\tan^2 x + \sec^2 x), \quad f''(0) = 1
$$

$$
f'''(x) = \sec(x) \tan(x) (\tan^2 x + \sec^2 x) + \sec(x) \left( 2 \tan x \sec^2 x + 2 \sec^3 x \tan x \right), \quad f'''(0) = 0
$$

$$
f^{(4)}(0) = 5
$$

| Derivada       | Valor |

| f'(0)      | 0     |
| f''(0)     | 1     |
| f'''(0)    | 0     |
| f''''(0)   | 5     |


## Série de taylor

Qual o valor de N da série de Taylor que será o seu limite? Por quê?

```
def secante(x):
    return 1/np.cos(x)

def taylor_secante(x, n):
    coef = [1, 1/2, 5/24, 61/720, 1385/40320] #par
    soma = np.zeros_like(x)
    for i in range(n):
        soma += coef[i]*x**(2*i)
    return soma

x = np.linspace(-1.5, 1.5, 200)

y_real = secante(x)
y2 = taylor_secante(x, 1)
y3 = taylor_secante(x, 2)
y4 = taylor_secante(x, 3)
y5 = taylor_secante(x, 4)

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(x, y_real, linewidth=2.0, color='lightblue', label='sec(x) exata')
ax.plot(x, y2, linewidth=1.0, color='black', linestyle='dashed', label='Taylor ordem 2')
ax.plot(x, y3, linewidth=1.0, color='red', linestyle='dashed', label='Taylor ordem 4')
ax.plot(x, y4, linewidth=1.0, color='green', linestyle='dashed', label='Taylor ordem 6')
ax.plot(x, y5, linewidth=1.0, color='purple', linestyle='dashed', label='Taylor ordem 8')

custom_ticks = [-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2]
custom_labels = [r'$-\pi/2$', r'$-\pi/4$', '0', r'$\pi/4$', r'$\pi/2$']
ax.set_xticks(custom_ticks)
ax.set_xticklabels(custom_labels)
ax.set_xlabel('x (radianos)')
ax.set_ylabel('sec(x)')
ax.set_title('Aproximação da secante pela série de Taylor')
ax.grid(True)
ax.legend()

plt.show()
```

Um gráfico mostrando a secante real vs séries de Taylor de diferentes ordens deixa claro:

Quanto maior
N, melhor a aproximação perto de 0

Mas todas explodem perto de +𝜋/2 e -π/2, mostrando o limite do intervalo de convergência.

## Aproximações e Propriedades da Função Secante

A função secante é definida como:

$$
\sec(x) = \frac{1}{\cos(x)}
$$

### ⚙️ Propriedades

- **Domínio:**  
  $$
  x \in \mathbb{R} \setminus \left\{ \frac{\pi}{2} + k\pi \mid k \in \mathbb{Z} \right\}
  $$  
  (a secante não está definida onde o cosseno é zero)

- **Paridade:**  
  $$
  \sec(-x) = \sec(x)
  $$  
  → **função par**

- **Período:**  
  $$
  2\pi
  $$  
  (igual ao do cosseno)

- **Assíntotas verticais:**  
  $$
  x = \frac{\pi}{2} + k\pi
  $$  
  onde a função tende a ±∞.

- **Relações trigonométricas úteis:**  
  $$
  \sec(x) = \frac{1}{\cos(x)} \quad \text{e} \quad \sec^2(x) = 1 + \tan^2(x)
  $$

---

### Aproximações (Série de Taylor em x = 0)

A série de Taylor da função secante em torno de \( x = 0 \) é:

$$
\sec(x) = 1 + \frac{x^2}{2} + \frac{5x^4}{24} + \frac{61x^6}{720} + \cdots
$$

- **Aproximação de ordem 0:**  
  $$
  \sec(x) \approx 1
  $$  
  — boa para valores de x muito próximos de 0.

- **Aproximação de ordem 2:**  
  $$
  \sec(x) \approx 1 + \frac{x^2}{2}
  $$  
  — melhora bastante a precisão perto da origem.

- **Aproximação de ordem 4:**  
  $$
  \sec(x) \approx 1 + \frac{x^2}{2} + \frac{5x^4}{24}
  $$  
  — aproxima bem até valores moderados de x.

## Código otimizado do secante

```
def secante_otimizada(x, n=5):
    numerador = 1
    denominador = 1
    soma_cos = 1
    sinal = -1
    for i in range(1, n):
        fator = 2 * i
        numerador *= x * x
        denominador *= fator * (fator - 1)
        soma_cos += sinal * numerador / denominador
        sinal *= -1
    return 1 / soma_cos
```

## Analise

```
repeticoes = 100
n_max = 10
enes = np.arange(1, n_max+1)
medias = np.zeros(n_max)
erros = np.zeros(n_max)
x = np.linspace(-np.pi/4, np.pi/4, 360)

def calcular_tempo_sec(x, n):
    inicio = timeit.default_timer()
    valor = secante_otimizada(x, n)
    erro = abs(valor - 1/np.cos(x))
    return timeit.default_timer() - inicio, erro

tempo_vector_sec = np.vectorize(calcular_tempo_sec)

for n in enes:
    tempos, erro = tempo_vector_sec(x, n)
    medias[n-1] = np.mean(tempos)*1000*1000
    erros[n-1] = np.sum(erro)/360
```

```
fig, ax = plt.subplots()
fig.set_size_inches(20, 8)

ax.plot(enes, medias, linewidth=2.0)
ax.grid()
ax.set_xlabel('N da Série de Taylor')
ax.set_ylabel('Tempo em μs', color="blue")
ax.tick_params(axis='y', labelcolor="blue")

ax2 = ax.twinx()
ax2.set_ylabel('Erro', color="red")
ax2.plot(enes, erros, color="red")
ax2.tick_params(axis='y', labelcolor="red")

plt.title('Tempo x Erro — Aproximação da Secante')
plt.show()
```
